## KID21

```python
light_dir = "D:/Data/LT218Chip1_BF_20221025_MIR8_5/12KIDs LN2 load - long/TD_Power"  # directory for the pulsedata
dark_dir = 'D:/Data/LT218Chip1_BF_20221025_MIR8_5/12KIDs mono off/TD_Power'  # directory for the noise data
LT = 218                # LT number of the chip. This is just for saving purposes
wavelength = 8500       # wavelenght of the laser. This is just for saving purposes
kid_nr = 21              # KID number
pread = 103            # readout power of the KID
comment = '20221025'    # the date of the experiment or any other comment you want to add for saving the figures
chuncksize = 40         # the number of binfiles that the code will analyse at once. The script will concatenate these files and analyze them at once istead of analyzing every file on its own
LT218_KID21_P103 = MKID(LT, wavelength, light_dir, dark_dir, kid_nr, pread, comment, chuncksize)
```

```python
mkid = LT218_KID21_P103      # choose the object you want to analyze
psc = 0.045
nsc = 0.045
settings = {                # Choose the best settings for the pulse analysis:
    'sf':1e6,'coord':'smith','response':'X','pw':256,'sw':30,'window':'box','ssf':1,'buffer':0.25,'mph':psc,'mpp':psc,'noise_mph':nsc,'noise_mpp':nsc,
    'nr_noise_segments':100,'binsize':0.02,'range':0.12,'fit_T':[50, 150],'max_bw':10000,'tlim':[0, 1],'filter_std':5,'rise_offset':0.1  
    } 
```

```python
# the mkid.initiate attribute is for quickly checking how the pulse selection and alignment is going. It will only analyze one chunck of binfiles that was set by the initiation of the object.
mkid.initiate(settings,f,binsize=0.1,dpulse=10,plot_pulse=5, below=.1)
```

```python
# the mkid.overview attribute does the full pulse analysis and provides an overview plot of all the relevant parameters
mkid.overview(
    settings,                       # input the pulse analysis settings                              
    f,                              # provide the required file with all the functions
    redo_peak_model=False,          # if you don't want to redo the pulse analysis but do want to renew the plot set this setting to False
    save=False,                      # if you want to save the figure and data, set it to True. It will create 4 files, a text file with the settings, a pickle file with the pulse data and a png and svg of file of the figure 
    figpath='./Results/'            # provide the directory to which the figure will be saved
    )
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
mkid.plot_mean_pulse(ax, 'lin')
```
